```python
from paddleocr import PaddleOCR, draw_ocr
from PIL import Image

ocr = PaddleOCR(use_angle_cls=True, lang='en')
img_path = '/content/received_1394536797852447.jpeg'

# Perform OCR
result = ocr.ocr(img_path, cls=True)

# Extract detected boxes and texts
boxes = [line[0] for line in result[0]]
txts = [line[1][0] for line in result[0]]

# Open the image
image = Image.open(img_path).convert('RGB')

# Draw bounding boxes on the image
im_show = draw_ocr(image, boxes, txts)

# Convert the image array to PIL Image
im_show = Image.fromarray(im_show)

# Save the image with bounding boxes drawn
im_show.save('result2.jpg')

# Save recognized text into a text file
output_file = 'recognized_text.txt'
with open(output_file, 'w') as file:
    for text in txts:
        file.write(text + '\n')
```
