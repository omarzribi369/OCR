```python
import pytesseract
import matplotlib.pyplot as plt
import cv2

# Read the image
im = cv2.imread("rd.jpg")
h, w, c = im.shape

# Set Tesseract OCR path
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"

# Perform OCR and get bounding boxes
boxes = pytesseract.image_to_boxes(im)
text  = pytesseract.image_to_string(im)
# Draw boxes on the image
for box in boxes.splitlines():
    box = box.split()
    x, y, x2, y2 = int(box[1]), int(box[2]), int(box[3]), int(box[4])
    cv2.rectangle(im, (x, h - y), (x2, h - y2), (0, 255, 0), 1)  # Draw green rectangles around text regions
print(text)
# Display the image with bounding boxes
plt.figure(figsize=(8, 6))
plt.imshow(cv2.cvtColor(im, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
